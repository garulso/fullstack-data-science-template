```python
# %%
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)
```

```python
# Carga ambos datasets
df_raw = pd.read_csv("../data/raw/titanic.csv")
df_clean = pd.read_csv("../data/processed/titanic_clean.csv")
```

```python
print(f"âœ… Crudo: {df_raw.shape}")
print(f"âœ… Limpio: {df_clean.shape}")
```

```python
# 1. Comparar nulos antes/despuÃ©s
nulls_raw = df_raw.isnull().sum()
nulls_clean = df_clean.isnull().sum()
```

```python
comparison = pd.DataFrame({
    "Crudo": nulls_raw,
    "Limpio": nulls_clean
})
print("ComparaciÃ³n de nulos:")
comparison[comparison["Crudo"] > 0]
```

```python
comparison["ReducciÃ³n (%)"] = (
    100 * (comparison["Crudo"] - comparison["Limpio"]) / comparison["Crudo"]
).round(1)
```

```python
print(comparison[comparison["Crudo"] > 0].to_string(index=True))
```

```python
# 2. Comparar distribuciÃ³n de Age (crudo vs limpio)
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
sns.histplot(df_raw["Age"], kde=True, bins=30, color="red", alpha=0.6)
plt.title("Edad â€” Datos crudos", fontsize=13)
plt.xlabel("Edad")
plt.ylabel("Conteo")

plt.subplot(1, 2, 2)
sns.histplot(df_clean["Age"], kde=True, bins=30, color="green", alpha=0.6)
plt.title("Edad â€” Datos limpios (mediana)", fontsize=13)
plt.xlabel("Edad")
plt.ylabel("Conteo")

plt.tight_layout()
plt.show()
```

```python
# 3. Ver impacto en anÃ¡lisis: supervivencia por edad (limpio)
plt.figure(figsize=(8, 4))
sns.boxplot(data=df_clean, x="Survived", y="Age", hue="Sex")
plt.title("Edad y supervivencia (dataset limpio)", fontsize=14)
plt.xlabel("Superviviente")
plt.ylabel("Edad")
plt.show()

print("ObservaciÃ³n: Ahora todas las edades estÃ¡n presentes â†’ anÃ¡lisis mÃ¡s completo.")
```

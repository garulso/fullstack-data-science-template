```python
# %%
# Carga el dataset LIMPIO (del sÃ¡bado)
df = pd.read_csv("../data/processed/titanic_clean.csv")
```

```python
print(f"âœ… Dataset limpio: {df.shape[0]} filas, {df.shape[1]} columnas")
df.head(3)
```

#### ***1. EstadÃ­sticas univariadas: numÃ©ricas***

```python
# Columnas numÃ©ricas clave
num_cols = ["Age", "Fare", "SibSp", "Parch"]
```

```python
# EstadÃ­sticas bÃ¡sicas con describe() â†’ pero vamos mÃ¡s allÃ¡
basic_stats = df[num_cols].describe()
basic_stats
```

```python
# AÃ±adimos varianza, asimetrÃ­a (skew) y curtosis
advanced_stats = pd.DataFrame({
    "media": df[num_cols].mean(),
    "mediana": df[num_cols].median(),
    "varianza": df[num_cols].var(),
    "desv_est": df[num_cols].std(),
    "asimetria": df[num_cols].skew(),
    "curtosis": df[num_cols].kurtosis()
})
advanced_stats.round(3)
```

#### ***2. EstadÃ­sticas por grupo: Â¿la clase afecta la edad?***

```python
# Media y mediana de Age por Pclass
age_by_class = df.groupby("Pclass")["Age"].agg(["mean", "median", "std", "count"])
age_by_class.round(2)
```

```python
# Â¿Y por supervivencia?
age_by_survival = df.groupby("Survived")["Age"].agg(["mean", "median", "std"])
age_by_survival.round(2)
```

```python
# Combinado: Pclass + Survived
age_by_both = df.groupby(["Pclass", "Survived"])["Age"].agg(["mean", "median", "count"])
age_by_both.round(2)
```

#### ***3. Cuartiles e IQR: base para detecciÃ³n de outliers***

```python
# Cuartiles de Age y Fare
Q1_age = df["Age"].quantile(0.25)
Q3_age = df["Age"].quantile(0.75)
IQR_age = Q3_age - Q1_age

Q1_fare = df["Fare"].quantile(0.25)
Q3_fare = df["Fare"].quantile(0.75)
IQR_fare = Q3_fare - Q1_fare
```

```python
print(f"Age â†’ Q1: {Q1_age:.1f}, Q3: {Q3_age:.1f}, IQR: {IQR_age:.1f}")
print(f"Fare â†’ Q1: {Q1_fare:.2f}, Q3: {Q3_fare:.2f}, IQR: {IQR_fare:.2f}")
```

```python
# LÃ­mites para outliers (regla IQR)
lower_age = Q1_age - 1.5 * IQR_age
upper_age = Q3_age + 1.5 * IQR_age

lower_fare = Q1_fare - 1.5 * IQR_fare
upper_fare = Q3_fare + 1.5 * IQR_fare
```

```python
print(f"\nOutliers en Age: <{lower_age:.1f} o >{upper_age:.1f}")
print(f"Outliers en Fare: <{lower_fare:.2f} o >{upper_fare:.2f}")
```
